# RAG + Agentic AI Exploration

This notebook demonstrates the RAG + Agentic AI system for data analysis.

```python
import sys
sys.path.append('../src')

from rag_pipeline import RAGPipeline, load_text_documents
from agent import DataAnalysisAgent
from config import settings
```

## 1. Initialize RAG Pipeline

```python
# Create RAG pipeline
rag = RAGPipeline()
print("✓ RAG Pipeline initialized")
```

## 2. Load and Ingest Documents

```python
# Load sample documents
try:
    documents = load_text_documents("../data/sample_data.txt")
    rag.ingest_documents(documents)
    print(f"✓ Loaded and ingested {len(documents)} documents")
except FileNotFoundError:
    print("Please create data/sample_data.txt first")
```

## 3. Test RAG Retrieval

```python
# Retrieve context for a query
query = "What is the company's revenue?"
context = rag.get_context(query)
print(f"Context for '{query}':\n{context}")
```

## 4. Initialize Agent

```python
# Create agent
agent = DataAnalysisAgent(rag)
print("✓ Agent initialized")
```

## 5. Run Analysis

```python
# Run analysis query
analysis_query = "Analyze the revenue trends and provide insights"
response = agent.analyze(analysis_query, context_query="company revenue")
print(f"Analysis Response:\n{response}")
```

## 6. Multi-step Analysis

```python
# Perform multiple analyses
queries = [
    "What are the main business drivers?",
    "Analyze customer satisfaction trends",
    "What recommendations would you make?"
]

results = agent.multi_step_analysis(queries)
for q, r in results.items():
    print(f"\nQuery: {q}")
    print(f"Response: {r}")
```

## 7. Conversation History

```python
# Display conversation history
history = agent.get_conversation_history()
print(f"Conversation History:\n{history}")
```
